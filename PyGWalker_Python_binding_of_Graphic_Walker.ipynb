# **PyGWalker "Python binding of Graphic Walker"**

PyGWalker can simplify your Jupyter Notebook data analysis and data visualization workflow, by turning your pandas dataframe into an interactive user interface for visual exploration.

It integrates Jupyter Notebook with Graphic Walker, an open-source alternative to Tableau. It allows data scientists to visualize / clean / annotates the data with simple drag-and-drop operations and even natural language queries.

Import pygwalker and pandas to your Notebook to get started.

```python
import pygwalker as pyg
import pandas as pd
```

The dataset contains hourly bike rental records from January 1, 2011, to December 31, 2012. Each row provides information for a specific hour and includes various attributes related to bike rentals. Here are the details of each column:

1. **date**: The date of the record in YYYY-MM-DD format.
2. **month**: The month of the year (1 to 12).
3. **season**: The season when the record was taken. Presumably encoded as:
   - 1: Winter
   - 2: Spring
   - 3: Summer
   - 4: Fall
4. **hour**: The hour of the day (0 to 23).
5. **year**: The year of the record.
6. **holiday**: Indicates whether the day is a holiday (1) or not (0).
7. **temperature**: The actual temperature recorded (in Celsius).
8. **feeling_temp**: The perceived temperature, also known as "feels like" temperature.
9. **humidity**: The humidity percentage.
10. **windspeed**: The wind speed recorded (presumably in meters per second).
11. **casual**: The number of casual (non-registered) bike rentals during that hour.
12. **registered**: The number of registered users who rented bikes during that hour.
13. **count**: The total number of bike rentals during that hour (casual + registered).
14. **work yes or not**: Indicates whether the day is a working day (1) or not (0).
15. **am or pm**: Indicates whether the time is AM or PM.
16. **Day of the week**: The day of the week (0 to 6), with the exact encoding typically being:
    - 0: Sunday
    - 1: Monday
    - 2: Tuesday
    - 3: Wednesday
    - 4: Thursday
    - 5: Friday
    - 6: Saturday



```python
df = pd.read_csv("https://kanaries-app.s3.ap-northeast-1.amazonaws.com/public-datasets/bike_sharing_dc.csv", parse_dates=['date'])
df
```

You can use pygwalker without changing your existing workflow. For example, you can call up Graphic Walker with the dataframe loaded in this way:

```python
pyg.walk(df)
```
